```python
f_m = 20e3 # for second plot/program f_m = 5e3
f_c = 100e3
N = 1e3
T = 1 /f_m
Fs = N/T
f = np.linspace(0,Fs/2,N/2)
t = np.linspace(0., T, N)

s_t = 0.5 * (1 - np.cos(2 * np.pi * f_m * t)) * np.sin(2 * np.pi * f_c * t)
plt.subplot(2, 1, 1)
plt.plot(t,s_t, 'blue')
plt.xlabel('Time')
plt.ylabel('s(t)')
plt.title('Time Domain')


fft_st = np.fft.fft(s_t)

plt.subplot(2, 1, 2)
plt.title('Fourier Transform')
plt.grid()
plt.xlabel('Frequency [kHz]')
plt.ylabel(r'S($\omega$)')
plt.plot(f*1e-3, np.absolute(fft_st[0:len(f)]), 'green')
plt.show()
```
